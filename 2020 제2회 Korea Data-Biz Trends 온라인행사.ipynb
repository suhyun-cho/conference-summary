### Session1. 상명대학교 소비자학과 (이준영 교수)
코로나19 이후 소비자행동 변화의 핵심은 온라인 소비의 증가, 그중에서도 시니어층의 온라인 소비가 증가했다는 점이 눈에 띈다.

* 한 예로, 화상 회의 서비스 zoom은 미국에 7개의 항공사보다도 더 잘나간다고 함.
* 현재 언택트 시대로 인해 VR이 주목받고 있음. (ex. Spatial이라는 스타트업에서 구현중인 VR 회의 사례)

#### 구독경제형
* 장점은, 기업이 고객과 지속적으로 관계를 유지해 나갈 수 있다는 점과 고객의 정보를 가져올 수 있다는 것.
    * ex) Netflix의 고객 취향 클러스터링

#### D2C (Direct to Consumer)
* 홈쇼핑 사례
* 스타벅스, 도미노피자 사례 모두 데이터기반 비즈니스 실행 중
    * ex) 피자 사진을 찍으면 피자를 주는 캠페인
* 맥도날드
    * 음성인식 회사와 제휴를 맺어서 고객이 자주 주문하는 메뉴를 추천해줌.
* cafeX
    * 직접 커피를 만들어주는 로봇 (잠실 롯데몰 등 여러곳에서 시범운행 했으나, 실제 커피맛이 좋지 않아서 점차 사라져 가는 중)
    * 본질적인 것이 중요하다는 것을 알려주는 사례
* Digital Trust
    * 고객은 데이터 안정성 문제에 불안해 하는데, 기업은 이를 해결할 방법을 고민해야 하는것이 과제.
* Digital Customer Experience
    * Customer Experience(CX) 는 Brand, Feel, Think, Relate, Act, Sense 이들의 조합.

### Session2. 한국 마이크로소프트 (오성미 이사)
일하는 방식의 변화 : Modern Work

조직내 협업을 데이터로 기록하고 대시보드화 한 Workplace Analytics Business Dashboards를 선보임.
> 업무 중 활용하는 데이터를 통합한 시스템. (ex. 원격으로 팀별 회의를 진행한 총 시간, 근무시간 등을 모두 데이터화 한 것)

* Q : 사람들의 잦은 질문중 하나는, 사람들이 원격 근무를 하면서 기록되는 모든 데이터가 추후 성과평가 같은것으로 반영되는거 아닌가?
    * A : 이에 대한 대답은 No. 개인을 식별할 수 없도록 함<br>
<br>

* 활용 사례
    * 예를들어, 어떤 팀 내에서 급격하게 소통이 늘었는데 사람 수는 그대로라면, 이는 capacity에 문제가 있는 것으로 판단하는 등으로 팀내 리소스 관리에도 활용.
    * 실제 분석 결과, 미팅을 짧게 끝내는 경우가 많음.

* 원격 근무로 인한 스트레스 지수 증가
    * 직원들이 개인 생활과 일의 분리가 안되는 문제를 지적함.
    * 실제로 뇌파지수 측정 결과 원격근무 회의 때 스트레스가 더 높게 나타났다고 함.

### 느낀점
기업들이 공통적으로 했던 말 중 하나가 코로나 시대를 예측하고 위에서 언급한 AI서비스를 개발한 것은 아니라는 말이었습니다. 어쩌면 코로나가 아니였어도 언젠가 우리에게 다가올 언택트 시대가 100걸음정도 빠르게 다가온게 아닌가 싶습니다. <br>

그럼에도 불구하고, 여전히 사람들이 우려하는 것은 사람과 사람간의 시너지입니다. 사람이 직접 얼굴을 보고 눈을 마주치면서 미팅을하고, 의견을 나누고, 서비스를 제공하면서 서로간에 느껴지는 온도(?) 감정(?) 같은 것은 기계로 대체하기 어려운 부분이라고 생각합니다. <br>

그런데, 이 또한 초반 정착시기에서 발생할 '불편함' 정도에서 그치고 이런 시스템에 사람이 적응하다 보면 우리가 우려하던 그것 또한 언젠가는 느껴지지 않을 만큼 편해질지도 모른다는 생각도 듭니다.
