* Cross-Validation
    * 모든 데이터는 동일한 모집단 분포에서 독립적으로 샘플링한 것으로 가정
    
**금융 시계열 데이터의 특성** <br>
Q : 금융 시계열 데이터도 independent and identically distributed로 가정할수있을까 ?<br>
* Non-Stationary Process - 시간 축에 따른 비정상성

**금융 투자 전략의 검증**<br>
> * 백테스팅
    * 과거 데이터에 시간 순서대로 설계한 모델에 의해 전략을 반복 실행하여 과거 일정 기간의 운용 수익을 추정하는 방법.
* 모의거래
    * 실제 일정 기간 만들어진 모델의 전략에 따라 모의 거래를 진행해 해당 기간의 운용 수익을 리스크 없이 측정하는 방법. <br>

**백테스팅 중 훈련한 모델의 검증**<br>
* 모멘텀 가정 : 시장은 상태나 추세를 유지하려는 경향이 있다.
* 유사 국면 가정 : 과거의 비슷한 추세를 따르는 경향.

**정규분포 가정** <br>
시장의 움직임은 정규분포를 따를 것이다 ?

* Fatter Tail & Taller Peak
    * 시장 참여자의 관심은 강한 가속력을 가지고 일부에 몰림
* 가설 검정, 프라이싱 모델, 리스크 모델에서 고려
    * 대부분의 가설 검정 과정은 정규분포를 가정함
    * Tail risk를 고려한 프라이싱과 리스크 모델링이 필요함
    * 이상치, 극단치가 학습 성능을 저해하므로
    
**Non-Stationary Process 가정**<br>
* 시장의 상태
    * 관측이 가능한 변수 : 지수, 가격, 거래량 등 정형화된 데이터
    * 관측이 어려운 변수 : 사람들의 관심과 기대심리, 향후 전망 등 비정형 데이터
* 시장의 상태에 대한 조건부 확률
    * 대부분의 독립 변수를 관측 가능하다면, 그 종속 변수인 시장 변화는 조건부 확률로 모델링 가능함.

## 4. LINE Wallet 추천 시스템
* CTR 2.5배 올린 이야기

김경민 / Naver clova

**내가 다음에 사용할 모듈은 UI상단으로 집합**
* 사용 확률에 따라서 UI상단에 배치
    * ex) 쿠폰 사용

**과거행동 기반 알고리즘**
* 월렛탭 행동로그(클릭/뷰) , 유사한 사용자 의 데이터를 수집
* Transformer알고리즘
    * 음성 인식, 합성, 번역기, 시계열, 이미지분류 등에 활용
    
**추천 시스템에서의 bias 문제**
> * 컨텐츠 내용과 관련없이 50%확률로 첫번째 모듈을 클릭한다.
* 첫번째 모듈이 클릭로그가 많았기 때문에 학습을 시키면, 다음에도 게속 상단에 배치됨.
* 해결방법
    * Loss function = cross entropy loss * module CTR 
        * (view에 대한 상대적 click수로 보정해주자)
    * view로 normalize가 되어서 상대적으로 클릭수가 많아도 view가 너무 많으면 penalty를 받게되는 형태
    * 결과 : 보정전CTR 22%, 보정 후 CTR 62% (약 3배증가)
    
**기획 vs 추천, 누가이길까**
* 사람이 직접 조정을 해서 떨어진 ctr을 올리기도 함.
* 서비스의 세로축을 바탕으로 추천함
    * ex) 유튜브와 유사.
    * 가로축으로 추천하는것도 필요하다는 것을 깨달음.

**일본 코로나 영향**
* 3/26 올림픽 연기하겠다고 발표한 날 쿠폰 사용자수가줄어듬.
    * 코로나로 외출이 줄어드니까
    * 찌라시 모듈은 감소함.
* But, 포인트 모듈은 증가함
    * 홈페이지 가입해서 어떤 미션을 수행하면 포인트를 얻는 형태.
* 이런상황에서 추천을 하게되면 포인트모듈이 상단에 배치됨.

**해결방안** 
> * 아이템 추천이 같이 이뤄져야 다른 모듈도 활성화된다고 판단.
* 가로축으로 쿠폰 추천을함.
* 그래프로 로그를 엮어라
    * 기존 머신러닝은 모든 로그가 독립적임
    * 최대 70일치 쿠폰 사용로그로 그래프를 만듬.
    * 사용자끼리도 서로 묶어줌.
* 그래프를 사용하게되면, 거리에 대한 힌트를 얻을 수 있음.
    * 인접한 노드끼리 representation을 섞어줌. 이를 몇번 반복하다보면 나와 떨어진 노드와도 representationd이 섞일 수 있음.

**그래프모델 장점**
* cold유저, cold 아이템에 대해서 더 잘 추천해줄 수 있음.
* 그래프로 묶여있으면, 아이템을 어떻게 추천해줄지에 대한 힌트를 줌.
* Top Popular추천대비 CTR 향상.

**얼마나 다양한 쿠폰을 추천해주나**
* 개인화척도
    * (0 : 모두 같은쿠폰, 1:모두 다른 쿠폰)
    * 그래프모델의 경우 0.85, 즉 다양하게 추천받는단 얘기
* 아이템 커버리지
    * 100개 아이템중 몇개를 추천을 보냈는가
    * 그래프모델의 경우 0.93으로 100개 아이템중 93개 아이템이 추천되었다는 뜻.
        * 비 그래프모델은 아이템의 2%만 추천이됨.
